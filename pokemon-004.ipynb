```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pokemon = pd.read_csv('pokemon_reduced.csv')
```

```python
pokemon.drop(['Number'], 1, inplace=True)
pokemon.drop(['Type 2'], 1, inplace=True)
pokemon.drop(['Generation'], 1, inplace=True)
pokemon.drop(['Legendary'], 1, inplace=True)
```

### Comparison

_Dratini_ vs _Wartortle_

| | |
|---------|------------|
| ![Dratini](http://testerstories.com/files/data_science/pokemon-Dratini.png) | ![Wartortle](http://testerstories.com/files/data_science/pokemon-Wartortle.png) |

```python
Pokemon1_name = 'Dratini'
Pokemon2_name = 'Wartortle'

Pokemon1 = pokemon.loc[pokemon['Name'] == Pokemon1_name]
Pokemon2 = pokemon.loc[pokemon['Name'] == Pokemon2_name]
```

```python
labels = 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'HP', 'Speed'

P1_data = []
P2_data = []

for i in labels:
    P1_data.append(int(Pokemon1[i]))

for i in labels:
    P2_data.append(int(Pokemon2[i]))
```

```python
index = P1_data.index(max(P1_data))

explode = [0, 0, 0, 0, 0]
explode.insert(index,0.1)

fig1 = plt.figure(figsize=(12, 8))
ax1 = fig1.add_subplot(111)

ax1.pie(P1_data, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') 
plt.title(Pokemon1_name + " Stats")
plt.show()
```

```python
index = P2_data.index(max(P2_data))

explode = [0, 0, 0, 0, 0]
explode.insert(index,0.1)


fig2 = plt.figure(figsize=(12, 8))
ax2 = fig2.add_subplot(111)

ax2.pie(P2_data, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax2.axis('equal') 
plt.title(Pokemon2_name + " Stats")
plt.show()
```

```python
ind = np.arange(6)
width = 0.25

fig = plt.figure( figsize=(12, 8) )
ax = fig.add_subplot(111)

ax.bar(ind, P1_data, width, color='r') 
ax.bar(ind + width, P2_data, width, color='b') 

ax.set_xticks(ind + width / 2)
ax.set_xticklabels(labels)
plt.title(Pokemon1_name + " vs " + Pokemon2_name)
plt.show()
```
